```
from sklearn.datasets import load_iris

iris_data=load_iris()
print(type(iris_data))
```

```
keys=iris_data.keys() #load한 data의 key들이 불러옴
print("붓꽃 데이터 세트의 키들:",keys)
```

**Key의 구성**<br>


*   data : feature의 data set
*   target : classification의 label값, regression의 numeric result
*   terget_names : label's name
*   featuer_names : feature's name
*   DESCR : explanation of data set and each feature

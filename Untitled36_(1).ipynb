##**1. 사용자 정의 함수 사용**

```

def split_train_test(data,test_ratio): 
  shuffled_indices=np.random.permutation(len(data)) #정수 배열을 인덱스로 생각하고 섞어주기
  test_set_size=int(len(data)*test_ratio)
  test_indices=shuffled_indices[:test_set_size]
  train_indices=shuffled_indices[test_set_size:]
  
  return data.iloc[train_indices],data.iloc[test_indices] # iloc : index 기반 indexing
```

```
train_set,test_set=split_train_test(housing,0.2)
print("train_set: ",len(train_set))
print("test_set: ",len(test_set))
```

```
#해시 함수
import hashlib
from zlib import crc32

def test_set_check(identifier,test_ratio,hash=hashlib.md5):
  return bytearray(hash(np.int64(identifier)).digest())[-1]<256*test_ratio

def split_train_test_by_id(data,test_ratio,id_column):
  ids=data[id_column]
  in_test_set=ids.apply(lambda id_: test_set_check(id_,test_ratio))
  return data.loc[~in_test_set],data.loc[in_test_set] #loc : 행의 선택
```

```
housing_with_id=housing.reset_index() # "index" column 추가
train_set,test_set=split_train_test_by_id(housing_with_id,0.2,"index")
```

```
housing_with_id["id"]=housing["longitude"]*1000+housing["latitude"] #경도와 위도를 가지고 id 만듦 -> 동일한 id가 나올 가능성 존재함
train_set,test_set=split_train_test_by_id(housing_with_id,0.2,"id") #그 식별자로 train과 test를 나눔
```

```
housing["income_cat"]=pd.cut(housing["median_income"], #pandas에서 cut으로 수치형 변수 구간을 나눠줌
                             bins=[0.,1.5,3.0,4.5,6.,np.inf], # 0~1.5, 1.5~3.0, 3.0~4.5, 4.5 ~ 6, 6~ 로 나누기
                             labels=[1,2,3,4,5]) #class값 차례로 1 2 3 4 5 부여
```

```
housing["income_cat"].value_counts() #income_cat이 범주형특성으로 바뀐 것을 알 수 있음
```

```
housing["income_cat"].hist() #median_income에서 왼쪽으로 편향된 그래프가 income_cat의 범주형 자료로 바뀌면서 표준형 분포 가짐
```

```
from sklearn.model_selection import StratifiedShuffleSplit #Stratified : 범주형 특성으로 고르게 분할 해주는 기능들 분포

split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index, test_index in split.split(housing,housing["income_cat"]): #고르게 나눌 범주형 특성을 지정해서 섞어서 분할 -> train/test index를 반환
  strat_train_set=housing.loc[train_index] # loc 이용 : 그 index에 맞는 data를 가져옴
  strat_test_set=housing.loc[test_index]
```

```
strat_test_set["income_cat"].value_counts()/len(strat_test_set)
```

```
housing["income_cat"].value_counts()/len(housing) #원래 데이터인 housing에서 income_cat의 train/test data에서 비율과도 동일
```

```
st_train_set, st_test_set=train_test_split(housing, test_size=0.2, random_state=42, stratify=housing["income_cat"])
#주로 sklearn의 stratify이용해서 해당 범주형 자료를 고르게 해서 나눠줌
```

```
st_train_set["income_cat"].value_counts()/len(st_train_set) #봐봡...비율 같지...??
```

```
def income_cat_proportions(data):
  return data["income_cat"].value_counts()/len(data)

train_set,test_set=train_test_split(housing,test_size=0.2,random_state=42)

compare_props=pd.DataFrame({
    "Overall" : income_cat_proportions(housing),
    "Stratified" : income_cat_proportions(strat_test_set),
    "Random" : income_cat_proportions(test_set),
}).sort_index()

compare_props["Rand.%error"]=100*compare_props["Random"]/compare_props["Overall"]-100
compare_props["Strat.%error"]=100*compare_props["Stratified"]/compare_props["Overall"]-100

#그냥 train_test_split 이랑 stratify로 해서 나눈 경우 비교
```
